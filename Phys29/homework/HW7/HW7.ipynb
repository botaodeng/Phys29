Problem 1

a)

since we have A is a constant, $\langle A\rangle = \int A P(x,y) dx dy = A \int P(x,y) dx dy$. Since P(x,y) is the distribution function, we have $ \int P(x,y) dxdy = 1 $. Thus we have $ \langle A\rangle = A \int P(x,y) dx dy = A * 1 = A $

b)

$ {\rm Var}(A) = \int [A - \langle A \rangle] ^2 P(x,y) dxdy = \int [A - A]^2 P(x,y)dxdy = \int 0 P(x,y)dxdy = 0 $

c)

$$

\langle Ax\rangle = \int Ax P(x,y)dxdy = A \int x P(x,y)dxdy = A \langle x\rangle

$$

$$

{\rm Var}(Ax) = \int [Ax - \langle Ax \rangle]^2 P(x,y) dxdy = \int [Ax -A\langle x\rangle]^2 P(x,y)dxdy = \int [A(x - \langle x\rangle)]^2 P(x,y)dxdy = \int A^2 [x-\langle x\rangle]^2 P(x,y) dxdy = A^2 \int [x - \langle x\rangle]^2 P(x,y) dxdy = A^2 {\rm Var}(x)

$$

Similarly, all these can be applied to ${\rm Var}(By)$, and we can get ${\rm Var}(By) = B^2 {\rm Var}(y)$

d)

$$

\langle Ax + By \rangle = \int (Ax + By) P(x,y) dxdy = \int (Ax)P(x,y) + (By)P(x,y) dxdy = \int (Ax)P(x,y) dxdy + \int (By)P(x,y) dxdy = A \int x P(x,y) dxdy + B \int y P(x,y) dxdy = A \langle x\rangle + B \langle y\rangle

$$

e)

$$

\langle x \rangle = \int x P(x,y) dxdy = \int x P(x)P(y) dxdy = \int x P(x) dx * \int P(y) dy = \int x P(x) dx * 1 = \int x P(x) dx

$$

Similarly we can probe $ \langle y \rangle = \int y P(y) dy $

f)

$$

{\rm Var}(x) = \int [x - \langle x\rangle]^2 P(x,y) dxdy = \int [x - \langle x\rangle]^2 P(x)P(y) dxdy = \int [x - \langle x\rangle]^2 P(x) dx * \int P(y) dy = \int [x - \langle x\rangle]^2 P(x) dx * 1 = \int [x - \langle x\rangle]^2 P(x) dx

$$

Similarly, we can prove $ {\rm Var}(y) = \int [y - \langle y\rangle]^2 P(y) dy $

g)

$$

{\rm Var}(Ax + By) = \int [(Ax + By) - \langle Ax + By\rangle]^2 P(x,y) dxdy = \int [Ax + By - A \langle x \rangle - B \langle y \rangle]^2 P(x,y) dxdy = \int [A(x - \langle x\rangle) + B(y - \langle y\rangle)]^2 P(x,y) dxdy

= \int [A(x - \langle x\rangle)]^2 P(x,y) + [B(y - \langle y\rangle)]^2 P(x,y) + 2AB(x - \langle x\rangle)(y - \langle y\rangle) P(x,y) dxdy \\

= \int A^2 [x - \langle x\rangle]^2 P(x,y) dxdy + \int B^2 [y - \langle y\rangle]^2 P(x,y) dxdy + 2AB \int (x - \langle x\rangle)(y - \langle y\rangle) P(x,y) dxdy \\

= A^2 \int [x - \langle x\rangle]^2 P(x) * \int P(y) dy + B^2 \int[y - \langle y\rangle]^2 P(y) dy * \int P(x) dx + 2AB \int (x - \langle x\rangle)P(x)dx * \int (y - \langle y\rangle) P(y) dy\\

= A^2 {\rm Var}(x) * 1 + B^2 {\rm Var}(y) * 1 + 2AB \int (x - \langle x\rangle)P(x)dx * \int (y - \langle y\rangle) P(y) dy

$$

since x and y is independed, $ \int (x - \langle x\rangle)P(x)dx * \int (y - \langle y\rangle) P(y) dy $ is 0.

Thus, we can get  $ {\rm Var}(Ax + By) = A^2{\rm Var}(x) + B^2{\rm Var}(y) $
